```python
# subplots() function takes three arguments rows, columns and plot number

# plt.subplot(1, 2, 1)
# The figure has 1 row, 2 columns, and this plot is the first plot.

# plt.subplot(1, 2, 2)
# the figure has 1 row, 2 columns, and this plot is the second plot.
```

```python
# row 1
# column 2

import matplotlib.pyplot as plt
import numpy
%matplotlib inline

# Plot 1
x = numpy.array([0, 1, 2, 3])
y = numpy.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x, y)

# Plot 2

a = numpy.array([0, 1, 2, 3])
b = numpy.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(a, b)

plt.show()
```

```python
# row 2
# column 1

# plot 1
plt.subplot(2, 1, 1)
plt.plot(x, y)

# plot 2
plt.subplot(2, 1, 2)
plt.plot(a, b)

plt.show()
```

```python
# row 2
# column 3

# plot 1
plt.subplot(2, 3, 1)
plt.plot(x, y)

# plot 2
plt.subplot(2, 3, 2)
plt.plot(a, b)

# plot 3
plt.subplot(2, 3, 3)
plt.plot(x, y)

# plot 4
plt.subplot(2, 3, 4)
plt.plot(a, b)

# plot 5
plt.subplot(2, 3, 5)
plt.plot(x, y)

# plot 6
plt.subplot(2, 3, 6)
plt.plot(a, b)

plt.show()
```

```python
# plt.tight_layout() function for fitting

# plot 1
plt.subplot(2, 3, 1)
plt.plot(x, y)

# plot 2
plt.subplot(2, 3, 2)
plt.plot(a, b)

# plot 3
plt.subplot(2, 3, 3)
plt.plot(x, y)

# plot 4
plt.subplot(2, 3, 4)
plt.plot(a, b)

# plot 5
plt.subplot(2, 3, 5)
plt.plot(x, y)

# plot 6
plt.subplot(2, 3, 6)
plt.plot(a, b)

plt.tight_layout()

plt.show()
```

```python
# row 1
# column 2
# Title : title(), suptitle()

# plot 1
plt.subplot(1, 2, 1)
plt.plot(a, b)
plt.title('Sales')

# plot 2
plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.title('Income')

plt.suptitle('My Shop')

plt.show()
```
