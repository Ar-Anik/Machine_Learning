```python
import pandas

data = pandas.read_csv('Buy Shirt.csv')

data
```

```python
x = data.drop(['Shirt_Buy'], axis='columns')

x
```

```python
y = data.Shirt_Buy

y
```

```python
# Encoding Data

from sklearn.preprocessing import LabelEncoder

x = x.apply(LabelEncoder().fit_transform)

x
```

```python
# Decision Tree Model Create

from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
```

```python
model.fit(x, y)
```

```python
# predict value

model.predict([[0, 2, 0, 1]])
```
