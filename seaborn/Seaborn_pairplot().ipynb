```python
# seaborn.pairplot(data, keyword)
# By pairplot plot multiple pairwise bivariate distributions in a dataset
# This shows the relationship for (n, 2) combination of variable in a DataFrame as a matrix of plots and the diagonal plots are the univariate plots.

import matplotlib.pyplot as plt
import numpy
import seaborn
%matplotlib inline
```

```python
penguins = seaborn.load_dataset('penguins')

penguins
```

```python
seaborn.pairplot(penguins)

seaborn.set()
```

```python
# hue keyword adds a semantic mapping 
# it changes the default marginal plot to a layered kernel density estimate

seaborn.pairplot(penguins, hue='island')
seaborn.set()
```

```python
# kind keyword
# Kind of plot to make.
# value are : {â€˜scatterâ€™, â€˜kdeâ€™, â€˜histâ€™, â€˜regâ€™}

seaborn.pairplot(penguins, kind = 'kde')
seaborn.set()
```

```python
seaborn.pairplot(penguins, kind='hist')
seaborn.set()
```

```python
seaborn.pairplot(penguins, kind='reg')
seaborn.set()
```

```python
# diag_kind
# kind of plot for the diagonal subplots.
# value are : {â€˜autoâ€™, â€˜histâ€™, â€˜kdeâ€™, None}

seaborn.pairplot(penguins, diag_kind = 'hist')
seaborn.set()
```

```python
seaborn.pairplot(penguins, diag_kind='kde')
seaborn.set()
```

```python
seaborn.pairplot(penguins, hue='species', diag_kind='hist')
seaborn.set()
```

```python
seaborn.pairplot(penguins, hue='species', markers=['o', 's', 'D'])
seaborn.set()
```

```python
# height keyword

seaborn.pairplot(penguins, height=1.5)
seaborn.set()
```

```python
seaborn.pairplot(penguins, height=2.5)
seaborn.set()
```

```python
seaborn.pairplot(penguins,
                 x_vars = ['bill_length_mm', 'bill_depth_mm', 'bill_depth_mm'],
                 y_vars = ['bill_depth_mm', 'flipper_length_mm', 'body_mass_g'],
                 )
seaborn.set()
```

```python
# corner keyword
# For plot only the lower triangle

seaborn.pairplot(penguins, corner=True)
seaborn.set()
```

```python
# plot_kws keyword

seaborn.pairplot(penguins, plot_kws=dict(marker='+'))
seaborn.set()
```

```python
seaborn.pairplot(penguins, plot_kws= dict(marker='+', linewidth=1))
seaborn.set()
```

```python
seaborn.pairplot(penguins, plot_kws=dict(marker='+', linewidth=3))
seaborn.set()
```

```python
# diag_kws keyword
# diag_kws are passed to the univariate plotting function

seaborn.pairplot(penguins, plot_kws=dict(marker='+', linewidth=2), diag_kws=dict(fill=False))
seaborn.set()
```

```python
seaborn.pairplot(penguins, plot_kws=dict(marker='*', linewidth=2), diag_kws=dict(bins=25, fill=False))
seaborn.set()
```
