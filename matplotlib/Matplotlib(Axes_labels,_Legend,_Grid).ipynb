```python
import matplotlib.pyplot as plt
%matplotlib inline

days = [1, 2, 3, 4, 5, 6, 7]

max_t = [58, 51, 52, 48, 47, 48, 46]

min_t = [43, 42, 40, 44, 33, 35, 37]

avg_t = [45, 48, 48, 46, 40, 42, 41]
```

```python
plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

```python
# add x axis label
# add y axis label
# add plot title

plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

```python
# legend() function indicate each plot label

plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label="Max_Temp.")
plt.plot(days, min_t, label="Min_Temp.")
plt.plot(days, avg_t, label="Avg_Temp.")

plt.legend()
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label="Max_Temp.")
plt.plot(days, min_t, label="Min_Temp.")
plt.plot(days, avg_t, label="Avg_Temp.")

plt.legend('ABC')
```

```python
# loc='best' use for bestfit legend

import matplotlib.pyplot as plt
%matplotlib inline

days = [1, 2, 3, 4, 5, 6, 7]
max_t = [32, 51, 52, 48, 47, 48, 46]
min_t = [43, 42, 40, 44, 33, 35, 37]
avg_t = [45, 48, 48, 46, 40, 42, 41]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label="Max_Temp.")
plt.plot(days, min_t, label="Min_Temp.")
plt.plot(days, avg_t, label="Avg_Temp.")

plt.legend(loc="best")
```

```python
# Shadow keyword

plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label="Max_Temp.")
plt.plot(days, min_t, label="Min_Temp.")
plt.plot(days, avg_t, label="Avg_Temp.")

plt.legend(loc="best", shadow=True)
```

```python
# fontsize keyword

plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label="Max_Temp.")
plt.plot(days, min_t, label="Min_Temp.")
plt.plot(days, avg_t, label="Avg_Temp.")

plt.legend(fontsize="small")
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label="Max_Temp.")
plt.plot(days, min_t, label="Min_Temp.")
plt.plot(days, avg_t, label="Avg_Temp.")

plt.legend(loc='lower center',fontsize="large")
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label="Max. Temp.")
plt.plot(days, min_t, label="Min. Temp.")
plt.plot(days, avg_t, label="Avg. Temp.")

plt.legend(loc='lower center', fontsize='x-small')
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label="Max. Temp.")
plt.plot(days, min_t, label="Min. Temp.")
plt.plot(days, avg_t, label="Avg. Temp.")

plt.legend(loc='lower center', fontsize='xx-small')
```

```python
# grid() function

plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label="Max_Temp.")
plt.plot(days, min_t, label="Min_Temp.")
plt.plot(days, avg_t, label="Avg_Temp.")

plt.legend(loc='best', shadow=True, fontsize='x-large')

plt.grid()
```
