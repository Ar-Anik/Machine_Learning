```
import pandas
import numpy

data = pandas.read_csv('/content/climate.csv')

data
```

```
new_data = data.replace(-99999, numpy.NaN)

new_data
```

```
new_data = data.replace([-99999, -88888], numpy.NaN)

new_data
```

```
new_data = data.replace({
    'Temperature' : -99999,
    'Windspeed' : [-99999, -88888],
    'Event' : 'No Event'
}, numpy.NaN)

new_data
```

```
new_data = data.replace({
    -99999 : numpy.NaN,
    -88888 : numpy.NaN,
    'No Event' : 'Sunny'
})

new_data
```

```
new_data = data.replace('[A-Za-z]', '', regex = True)

new_data
```

```
new_data = data.replace({
    'Temperature' : '[A-Za-z]',
    'Windspeed' : '[A-Za-z]',
}, '', regex = True)

new_data
```

```
data = pandas.DataFrame({
    'Score' : ['Exceptional', 'Average', 'Good', 'Poor', 'Average', 'Exceptional'],
    'Student' : ['Anik', 'Joy', 'Oni', 'Tom', 'Bom', 'Rob'],
})

data
```

```
new_data = data.replace(['Poor', 'Good', 'Average', 'Exceptional'], [1, 2, 3, 4])

new_data
```
