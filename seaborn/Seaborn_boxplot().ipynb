```python
import matplotlib.pyplot as plt
import seaborn
import pandas
%matplotlib inline
```

```python
penguins = seaborn.load_dataset('penguins')

penguins
```

```python
penguins.head()
```

```python
penguins.describe()
```

```python
# boxplot() show min, 25%, 50%, 75%, max.

seaborn.boxplot(penguins.bill_length_mm)
seaborn.set()
```

```python
seaborn.boxplot(x='island', y='bill_length_mm', data=penguins)
seaborn.set()
```

```python
# hue keyword

seaborn.boxplot(x='island', y='bill_length_mm', hue = 'sex', data=penguins)
seaborn.set()
```

```python
# axis change

seaborn.boxplot(x='bill_length_mm', y='island', hue='sex', data=penguins)
seaborn.set()
```

```python
# ordering of bars

seaborn.boxplot(x='bill_length_mm', y='island', hue='sex', order = ['Torgersen', 'Dream', 'Biscoe'], data=penguins)
seaborn.set()
```

```python
# color keyword

seaborn.boxplot(x='island', y='bill_length_mm', color = 'orange', data=penguins)
seaborn.set()
```

```python
seaborn.boxplot(x='island', y='bill_length_mm', hue='sex', color="orange", data=penguins)
seaborn.set()
```

```python
seaborn.boxplot(x='island', y='bill_length_mm', color="0.25", data=penguins)
seaborn.set()
```

```python
seaborn.boxplot(x='island', y='bill_length_mm', color="0.75", data=penguins)
seaborn.set()
```

```python
# width keyword

seaborn.boxplot(x='island', y='bill_length_mm', hue='sex', width = 0.5, data=penguins)
seaborn.set()
```

```python
seaborn.boxplot(x='island', y='bill_length_mm', hue='sex', width=3, data=penguins)
seaborn.set()
```

```python
# whis keyword
# Points outside this range will be identified as outliers

seaborn.boxplot(x='island', y='bill_length_mm', hue='sex', whis = 0.5, data=penguins)
seaborn.set()
```

```python
seaborn.boxplot(x='island', y='bill_length_mm', hue='sex', whis = 0.2, data=penguins)
seaborn.set()
```

```python
# showcaps keyword
# Default its True

seaborn.boxplot(x='island', y='bill_length_mm', showcaps=False, data=penguins)
seaborn.set()
```

```python
# catplot()
# for join two boxplot

tips = seaborn.load_dataset('tips')

seaborn.catplot(x='sex', y='total_bill', hue='smoker', col='time', kind='box', data=tips)
seaborn.set()
```
