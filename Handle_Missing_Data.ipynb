```
 # fillna to fill missing values using different ways
# interpolate to make a guess on missing values using interpolation
# dropna to drop rows with missing values
```

```
import pandas

data = pandas.read_csv('/content/weather.csv')

data
```

```
# Data type of Day Column

type(data.Day[0])
```

```
# Day column data are string type it convert to times

data = pandas.read_csv('/content/weather.csv', parse_dates = ['Day'])

data
```

```
type(data.Day[0])
```

```
# Set index as a Day Column

data.set_index('Day', inplace = True)

data
```

```
# all NaN Value replace to 0.0

new_data = data.fillna(0)

new_data
```

```
new_data = data.fillna({
    'Temperature' : 0,
    'Windspeed' : 0,
    'Event' : "No Event"
})

new_data
```

```
# Forward Fill value next empty value

new_data = data.fillna(method="ffill")

new_data
```

```
# backward fill value for upper empty value

new_data = data.fillna(method="bfill")

new_data
```

```
new_data = data.fillna(method="pad")

new_data
```

```
new_data = data.fillna(method = "ffill", axis="columns")

new_data
```

```
new_data = data.fillna(method = "bfill", axis="columns")

new_data
```

```
data1 = data.fillna(method = "ffill")

#forward fill only one time

data2 = data.fillna(method = "ffill", limit=1)

print(data1, "\n\nNew Data Frame\n\n", data2)
```

```
# empty cell are fill upper and lower near value
new_data = data.interpolate()

new_data
```

```
# When empty cell are fill with consider their date

new_data = data.interpolate(method="time")

new_data
```

```
# All full-fill row

new_data = data.dropna()

new_data
```

```
# All column not fill-up row are missing

new_data = data.dropna(how="all")

new_data
```

```
# Another way for All column not fill-up row are missing

new_data = data.dropna(thresh=1)

new_data
```

```
new_data = data.dropna(thresh=2)

new_data
```

```
new_data = data.dropna(thresh=3)

new_data
```

```
# want to show all missing date

ranges = pandas.date_range('2021-01-01', '2021-01-11')

indx = pandas.DatetimeIndex(ranges)

new_data = data.reindex(indx)

new_data
```
