```python
# Change Event row NAN value to Sunny  

def change_Event_cell(value) :
  if value == "Not Available" :
    return "Sunny"
  return value

def change_Temperature_cell(value) :
  if value == "Not Available" :
    return None
  return value

def change_Windspeed_cell(value) :
  if value == "Not Available" :
    return None
  return value

data = pandas.read_excel('/content/Temperature.xlsx', "Sheet1", converters = {
    'Event' : change_Event_cell,
    'Windspeed' : change_Windspeed_cell,
    'Temperature' : change_Temperature_cell
})

data
```

```python
data.to_excel("new.xlsx", sheet_name="weather", index=False)

read = pandas.read_excel('/content/new.xlsx')
read
```

```python
data.to_excel("new_excel.xlsx", sheet_name="weather", startrow= 2, startcol= 2, index=False)

read = pandas.read_excel('/content/new_excel.xlsx')
read
```

```python
data_stock = pandas.DataFrame({
    'Tickers' : ['GOOGLE', 'WMT', 'MSFT'],
    'Price' : [845, 65, 64],
    'People' : ['Larry Page', 'Sam Walton', 'Bill Gates'],
    'EPS' : [27.82, 4.61, 2.12],
})

data_weather = pandas.DataFrame({
    'Day' : ['19/6/2021', '20/6/2021', '21/6/2021'],
    'Temperature' : [32, 35, 28],
    'Event' : ['Rain', 'Sunny', 'Snow']
})

print(data_stock, '\n\n', data_weather)
```

```python
with pandas.ExcelWriter('stocks_weather.xlsx') as writer :
   data_stock.to_excel(writer, sheet_name="stock", index = False)
   data_weather.to_excel(writer, sheet_name="weather", index = False)
```
