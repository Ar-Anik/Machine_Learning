```python
# Gradient Descent is an algorithm that finds best fit line for given training Data set

import numpy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Gradient Descent Function

def gradient_descent(x, y) :
  m = 0
  b = 0

  iterations = 100
  n = len(x)

  learning_rate = 0.08

  for i in range(iterations) :
    y_predict = m * x + b

    cost = (1/n) * sum([val*val for val in (y_predict - y)])

    m_diff = (2/n) * sum(x * (y_predict - y))
    b_diff = (2/n) * sum(y_predict - y)

    m = m - (learning_rate * m_diff)
    b = b - (learning_rate * b_diff)

    print("m = {}, b = {}, cost = {}, iteration = {}\n".format(m, b, cost, i))

    plt.plot(x, y_predict)
```

```python
x = numpy.array([1, 2, 3, 4, 5])
y = numpy.array([5, 7, 9, 11, 13])

gradient_descent(x, y)
```
