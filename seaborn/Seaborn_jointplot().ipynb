```python
import matplotlib.pyplot as plt
import numpy
import seaborn
%matplotlib inline
```

```python
penguins = seaborn.load_dataset('penguins')

penguins
```

```python
seaborn.jointplot(x='bill_length_mm', y='bill_depth_mm', data=penguins)

seaborn.set()
```

```python
# hue keyword will add conditional colors to the scatterplot
# draw separate density curves (using kdeplot()) on the marginal axes

seaborn.jointplot(x='bill_length_mm', y='bill_depth_mm', hue = 'island', data=penguins)

seaborn.set()
```

```python
seaborn.jointplot(x='bill_length_mm', y='bill_depth_mm', hue = 'sex', data=penguins)

seaborn.set()
```

```python
# kind keyword 
# several different approaches to plotting are available through the kind parameter
# bydefault kind keyword value is 'scatter'

# kind = 'kde' draw both bivariate and univariate

seaborn.jointplot(x='bill_length_mm', y='bill_depth_mm', kind = 'kde', data=penguins)

seaborn.set()
```

```python
# kind = 'hist' draw histogram

seaborn.jointplot(x='bill_length_mm', y='bill_depth_mm', kind = 'hist', data=penguins)

seaborn.set()
```

```python
# kind = 'hex'

seaborn.jointplot(x='bill_length_mm', y='bill_depth_mm', kind = 'hex', data=penguins)

seaborn.set()
```

```python
# kind="reg" to add a linear regression fit and univariate kde curves

seaborn.jointplot(x='flipper_length_mm', y='body_mass_g', kind = 'reg', data=penguins)

seaborn.set()
```

```python
# kind = resid

seaborn.jointplot(x='flipper_length_mm', y='body_mass_g', kind = 'resid', data=penguins)

seaborn.set()
```

```python
# marker keyword

seaborn.jointplot(x='flipper_length_mm', y='body_mass_g', marker = '+', data=penguins)

seaborn.set()
```

```python
# marginal_kws is keyword arguments

seaborn.jointplot(x='flipper_length_mm', y='body_mass_g', marker='*', marginal_kws = dict(fill=False), data=penguins)

seaborn.set()
```

```python
seaborn.jointplot(x='flipper_length_mm', y='body_mass_g', marker='*', marginal_kws = dict(bins=25, fill=False), data=penguins)

seaborn.set()
```

```python
# height = value
# height keyword define size of the figure

seaborn.jointplot(x='bill_length_mm', y='bill_depth_mm', height=5, data=penguins)

seaborn.set()
```

```python
seaborn.jointplot(x='bill_length_mm', y='bill_depth_mm', height=10, data=penguins)

seaborn.set()
```

```python
# ratio keyword
# ratio of joint axes height to marginal axes height.

seaborn.jointplot(x='bill_length_mm', y='bill_depth_mm', ratio = 2, data=penguins)
seaborn.set()
```

```python
seaborn.jointplot(x='bill_length_mm', y='bill_depth_mm', ratio = 5, data=penguins)
seaborn.set()
```

```python
seaborn.jointplot(x='bill_length_mm', y='bill_depth_mm', ratio = 10, data=penguins)
seaborn.set()
```

```python
# space keyword
# Space between the joint and marginal axes
# space value 0 to 100

seaborn.jointplot(x='bill_length_mm', y='bill_depth_mm', space = 0.1, data=penguins)
seaborn.set()
```

```python
seaborn.jointplot(x='bill_length_mm', y='bill_depth_mm', space = 0.9, data=penguins)
seaborn.set()
```

```python
seaborn.jointplot(x='bill_length_mm', y='bill_depth_mm', space = 5, data=penguins)
seaborn.set()
```

```python
# color keyword

seaborn.jointplot(x='bill_length_mm', y='bill_depth_mm', color='orange', data=penguins)
seaborn.set()
```

```python
# marginal_ticks keyword

seaborn.jointplot(x="bill_length_mm", y="bill_depth_mm", height=8, ratio=5, marginal_ticks = True, data=penguins)
seaborn.set()
```

```python
seaborn.jointplot(x='bill_length_mm', y='bill_depth_mm', height=8, ratio=5, marginal_ticks = False, data=penguins)
seaborn.set()
```
