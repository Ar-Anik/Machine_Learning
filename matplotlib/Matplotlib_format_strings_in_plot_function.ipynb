```python
import matplotlib.pyplot as plt
%matplotlib inline

x = [1, 2, 3, 4, 5, 6, 7]
y = [50, 51, 52, 48, 47, 48, 46]

plt.plot(x, y)
```

```python
# we change graph color blue to green

plt.plot(x, y, 'g')
```

```python
plt.plot(x, y, 'c')
```

```python
# we cann't see whole graph we want to see each point by '+' symbol

plt.plot(x, y, 'g+')
```

```python
plt.plot(x, y, 'r--')
```

```python
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

plt.plot(x, y, 'k+-.')
```

```python
plt.plot(x, y, 'or')
```

```python
plt.plot(x, y, 'D')
```

```python
plt.plot(x, y, 'H')
```

```python
plt.plot(x, y, '4')
```

```python
plt.plot(x, y, ':')
```

```python
plt.plot(x, y, '^k')
```

```python
# In format strings order doesn't matter

plt.plot(x, y, '--y*')
```

```python
plt.plot(x, y, 'bD--')
```

```python
plt.plot(x, y, color='blue', marker='D', linestyle='--')
```

```python
plt.plot(x, y, color='#008840', marker='*', markersize=20, linestyle='')
```

```python
# alpha value set the visibility of a graph alpha value 0 to 1

plt.plot(x, y, color='#008840', alpha=0)
```

```python
plt.plot(x, y, color='#008840', alpha=0.2)
```
