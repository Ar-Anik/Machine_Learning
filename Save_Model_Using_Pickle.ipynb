Pickle is the standard way of serializing objects in Python.

Pickle operation are for serialize machine learning algorithms and save the serialized format to a file.

```python
import pandas
import numpy
from sklearn import linear_model
```

```python
data = pandas.read_csv('HomePrice.csv')

data
```

```python
# Create Linear Regression Model and Train that Model

model = linear_model.LinearRegression()
model.fit(data[['Area']], data.Price)
```

```python
# We know linear hypothesis is y = m*x + b
# here co-efficient of x is m
# b is intercept

# the value of m
model.coef_
```

```python
# the value of intercept

model.intercept_
```

```python
# predict price of a area by linear model

model.predict([[4500]])
```

```python
# save model to a file using python pickle
# first we create and open a file in binary write mode
# then linear model copy into that file

import pickle

with open('pickle_file', 'wb') as file : 
  pickle.dump(model, file)
```

```python
# load save model by pickle.load() function

with open('pickle_file', 'rb') as file :
  pk_model = pickle.load(file)
```

```python
# now do same thing by pickle_model that we do by Linear Regression model

pk_model.coef_
```

```python
pk_model.intercept_
```

```python
pk_model.predict([[4500]])
```
