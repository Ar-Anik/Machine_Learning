```python
# distribution plots in seaborn which is used for examining univariate and bivariate distributions
# 4 types of distribution plots : 
# 1. joinplot
# 2. distplot
# 3. pairplot
# 4. rugplot
```

```python
 import numpy
 import seaborn
 import pandas
 import matplotlib.pyplot as plt
```

```python
seaborn.get_dataset_names()
```

```python
penguins = seaborn.load_dataset('penguins')

penguins
```

```python
penguins.head()
```

```python
# Displots used basically for univariant set of observations and visualizes it through a histogram

seaborn.distplot(penguins.flipper_length_mm)
```

```python
# hist keyword By default it is true
# hist keyword when false histrogram is delete

seaborn.distplot(penguins.flipper_length_mm, hist = False)
```

```python
# rug keyword for rug plot

seaborn.distplot(penguins.flipper_length_mm, rug = True)
```

```python
# by bins keyword along x-axis bins are set

seaborn.distplot(penguins.flipper_length_mm, bins = 20)
```

```python
# bins set by list

seaborn.distplot(penguins.flipper_length_mm, bins = [180, 190, 200, 230, 240])
```

```python
# kde keyword if it false, delete plot of gaussian kernel density

seaborn.distplot(penguins.flipper_length_mm, kde = False)
```

```python
# norm_hist if it True, the histogram height shows a density rather than a count. This is implied if a KDE or fitted density is plotted.

# Normal distplot
seaborn.distplot(penguins.flipper_length_mm)
```

```python
seaborn.distplot(penguins.flipper_length_mm, hist = True, kde = False)
```

```python
# styling rugplot

seaborn.distplot(penguins.flipper_length_mm, rug_kws = {'color' : 'red', 'height' : 0.3}, rug = True)
```

```python
# styling histograms

seaborn.distplot(penguins.flipper_length_mm, kde = False, hist_kws = {'color' : 'orange', 'histtype' : 'step'})
```

```python
seaborn.distplot(penguins.flipper_length_mm, kde=False, hist_kws = {'color' : 'orange', 'alpha' : 0.8, 'histtype' : 'step'})
```

```python
seaborn.distplot(penguins.flipper_length_mm, kde=False, hist_kws = {'color' : 'orange', 'alpha' : 0.6 ,'histtype' : 'stepfilled'})
```

```python
seaborn.distplot(penguins.flipper_length_mm, kde=False, hist_kws={'color' : 'orange', 'alpha' : 1, 'histtype' : 'stepfilled'})
```

```python
# styling KDE Plots

seaborn.distplot(penguins.flipper_length_mm, hist=False, kde_kws = {'color' : 'orange', 'alpha' : 0.8})
```

```python
# lw keyword define density of kde plot

seaborn.distplot(penguins.flipper_length_mm, hist=False, kde_kws = {'color' : 'orange', 'lw' : 1, 'alpha' : 0.9})
```

```python
seaborn.distplot(penguins.flipper_length_mm, hist=False, kde_kws={'color' : 'orange', 'lw' : 2, 'alpha' : 0.9})
```

```python
seaborn.distplot(penguins.flipper_length_mm, hist=False, kde_kws = {'color' : 'orange', 'lw' : 3, 'alpha' : 0.9})
```
