**Categorical Variable** : The Values of a categorical variable are selected from a
group of categories, also called labels.

Two Type of Categoriacal Variables : <br>
1) Ordinal <br>
2) Nominal

**Ordinal Variables** : Ordinal Variables are categorical variable in which the categories can be meaningfully ordered.

When the ordinal variable convert into the number then it has mathematical value.

Example : Rating of Product

**Nominal Varialbe** : Nominal variable is same like ordinal variable but there is nothing that indicates an intrinsic order of the labels and in principle.

All labels are equal no order exist.

No mathematical value.

Example : Colour(Red, Green, White, Black)



Machines understand numbers, not text. We need to convert each text category to numbers in order for the machine to process them using mathematical equations. So we need Encoding.


**Q** : What is Categorical Encoding? <br>
Any Structured dataset includes multiple columns and columns are combination of numerical as well as categorical variables. A machine can only understand the numbers. It cannot understand the text. For this reason we need to convert categorical columns to numerical columns so that machine learning algorithm understands it. This process is called categorical encoding.

***Categorical encoding is a process of converting categories to numbers.***

P1.jpeg

[For More Details](https://https://www.analyticsvidhya.com/blog/2020/03/one-hot-encoding-vs-label-encoding-using-scikit-learn/)

Two Popular Approaches to Categorical Encoding Are : <br>
1) Label Encoding <br>
2) One-Hot Encoding <br>

***Label Encoding*** : Label Encoding is a popular encoding technique for handling categorical variables. In this technique, each label is assigned a unique integer based on alphabetical ordering.

```python
import pandas
import numpy

data = pandas.read_csv('Salary.csv')

data
```

```python
# Data info print

data.info
```

```python
# Above dataframe, Country column data are categorical nominal data
# so we convert this column into numerical data

from sklearn import preprocessing

# from  preprocessing we create Label Encoder object
encoder_object = preprocessing.LabelEncoder()

# Encode Country Column
data['Country'] = encoder_object.fit_transform(data['Country'])
```

```python
# alphabetically numbering categorical value to numeric value 

data
```

```python
# Now Linear Regression
from sklearn import linear_model

r = linear_model.LinearRegression()

r.fit(data[['Country', 'Age']], data.Salary)
```

```python
r.coef_
```

```python
r.intercept_
```

```python
# predict salary which country = Bangladesh(0) and age = 18

r.predict([[0, 18]])
```

```python
# predict salary which country = China(1) and age = 18

r.predict([[1, 18]])
```
