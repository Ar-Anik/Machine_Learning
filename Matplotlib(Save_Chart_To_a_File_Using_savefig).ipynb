```python
import matplotlib.pyplot as plt
%matplotlib inline

array = [20, 25, 15, 10, 30]
mylabel = ['Math', 'Bangla', 'English', 'Biology', 'Chemistry']
mycolor = ['black', 'Red', 'Green', 'Cyan', 'Orange']
```

```python
plt.pie(array, labels=mylabel, colors=mycolor, autopct='%0.2f%%', explode=[0, 0.1, 0, 0, 0.2])

plt.axis("equal")
plt.show()
```

```python
# savefig() function for save pie-chart as pic, pdf etc format

plt.axis("equal")
plt.pie(array, labels=mylabel, colors=mycolor, autopct='%0.2f%%', explode=[0, 0.1, 0, 0, 0.2])

plt.savefig('pie-chart.png')
```

```python
# bbox_inches keyword use for fit the png file

plt.axis("equal")
plt.pie(array, labels=mylabel, colors=mycolor, autopct='%0.2f%%', explode=[0, 0.1, 0, 0, 0.2])

plt.savefig('pie-chart.png', bbox_inches='tight')
```

```python
# Transparent keyword use for background change

plt.axis("equal")
plt.pie(array, labels=mylabel, colors=mycolor, autopct='%0.2f%%', explode=[0, 0.1, 0, 0, 0.2])

plt.savefig('pie-chart.png', bbox_inches='tight', transparent=True)
```

```python
# pad_inches keyword use for padding in picture

plt.axis("equal")
plt.pie(array, labels=mylabel, colors=mycolor, autopct='%0.1f%%', explode=[0.1, 0, 0, 0.3, 0.1])

plt.savefig('pie-chart', bbox_inches="tight", pad_inches=2, transparent=True)
```

```python
# save pie-chart in specific location

plt.axis("equal")
plt.pie(array, labels=mylabel, colors=mycolor, autopct='%0.1f%%', explode=[0.2, 0, 0, 0.1, 0.2])

plt.savefig('I:/Python/PythonCode/Machine Learning/pie-chart.pdf', bbox_inches="tight", pad_inches=2, transparent=True)        # it work only jupeter, anaconda, computer exist ide
```
