```
# pyplot is a plot object in matplotlib modulo
# The inline option with the %matplotlib magic function renders the plot out cell even if show() function of plot object is not called.

import matplotlib.pyplot as plt
%matplotlib inline 

x = [1, 2, 3, 4, 5, 6, 7]
y = [50, 51, 52, 48, 47, 49, 46]

plt.plot(x, y)
```

```
# Color Change of Plot

plt.plot(x, y, color='green')
```

```
plt.plot(x, y, color='red')
```

```
plt.plot(x, y, color='yellow')
```

```
# Linewidth increase

plt.plot(x, y, linewidth=6)
```

```
# Linestyle Keyword

plt.plot(x, y, linestyle='dashed')
```

```
plt.plot(x, y, linestyle='dashdot')
```

```
plt.plot(x, y, linestyle='dotted')
```
