```python
# scatter() function plots one dot for each observation.
# it needs two arrays of the same length

import matplotlib.pyplot as plt
import numpy
%matplotlib inline
```

```python
# The X-axis shows how old the car is.
# The Y-axis shows the speed of the car when it passes.

x = numpy.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = numpy.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])

a = numpy.array([2, 2, 8, 1, 15, 8, 12, 9, 7, 3, 11, 4, 7, 14, 12])
b = numpy.array([100, 105, 84, 105, 90, 99, 90, 95, 94, 100, 79, 112, 91, 80, 85])
```

```python
plt.scatter(x, y)

plt.show()
```

```python
# Draw two plot on the same figure

plt.scatter(x, y)

plt.scatter(a, b)

plt.show()
```

```python
# color keyword

plt.scatter(x, y, color = 'orange')

plt.scatter(a, b, color = 'black')

plt.show()
```

```python
# Each Dot Color

x_arr = [1, 2, 3, 4, 5, 6]
y_arr = [5, 10, 15, 20, 25, 30]

dot_color = ['black', 'orange', 'red', 'green', 'magenta', 'red']

plt.scatter(x_arr, y_arr, color = dot_color)

plt.show()
```

```python
plt.scatter(x_arr, y_arr, c = dot_color)

plt.show()
```

```python
# ColorMap = cmap
# colormap is list of colors, where each color has a value that ranges from 0 to 100
# colormap is called 'viridis'

x_arr = [1, 2, 3, 4, 5, 6]
y_arr = [5, 10, 15, 20, 25, 30]

dot_color = [10, 20, 30, 40, 50, 60]

plt.scatter(x_arr, y_arr, c=dot_color, cmap = 'viridis')

plt.show()
```

```python
# include actual colormap by colorbar() function

x_arr = [1, 2, 3, 4, 5, 6]
y_arr = [5, 10, 15, 20, 25, 30]
dot_color = [10, 20, 30, 40, 50, 60]

plt.scatter(x_arr, y_arr, c=dot_color, cmap='viridis')

plt.colorbar()

plt.show()
```

```python
# size of each dot

x_arr = [1, 2, 3, 4, 5, 6]
y_arr = [5, 10, 15, 20, 25, 30]

dot_size = [100, 200, 300, 400, 500, 1000]

plt.scatter(x_arr, y_arr, c='orange', s = dot_size)

plt.show()
```

```python
# Dot color density Change by alpha keyword it value 0 to 1

x_arr = [1, 2, 3, 4, 5, 6]
y_arr = [5, 10, 15, 20, 25, 30]

dot_size = [100, 200, 300, 400, 500, 1000]

plt.scatter(x_arr, y_arr, c='orange', s=dot_size, alpha = 0.7)

plt.show()
```

```python
x_arr = [1, 2, 3, 4, 5, 6]
y_arr = [5, 10, 15, 20, 25, 30]

dot_size = [100, 200, 300, 400, 500, 1000]

plt.scatter(x_arr, y_arr, c='orange', s=dot_size, alpha = 0.3)

plt.show()
```

```python
x = numpy.random.randint(100, size=100)

y = numpy.random.randint(100, size=100)

colors = numpy.random.randint(100, size=100)

size = 10 * x

plt.scatter(x, y, c=colors, s=size, alpha=0.6, cmap = 'nipy_spectral_r')

plt.colorbar()

plt.show()
```

```python
plt.scatter(x, y, c=colors, s=size, alpha=0.5, cmap='prism')

plt.show()
```

```python
plt.scatter(x, y, c=colors, s=size, alpha=0.7, cmap='cubehelix')

plt.show()
```

```python
plt.scatter(x, y, c=colors, s=size, alpha=0.8, cmap='summer')

plt.show()
```
