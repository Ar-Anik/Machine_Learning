```python
import matplotlib.pyplot as plt
import numpy
%matplotlib inline

arr = numpy.array([35, 25, 25, 15])

plt.pie(arr)
```

```python
expense_money = [1400, 600, 300, 410, 250]
expense_labels = ['Home Rent', 'Food', 'Phone/Internet Bill', 'Car', 'Other Utilities']
```

```python
plt.pie(expense_money, labels=expense_labels)
```

```python
# axis() fun

plt.axis("equal")
plt.pie(expense_money, labels=expense_labels)
```

```python
# show() function

plt.axis("equal")
plt.pie(expense_money, labels=expense_labels)

plt.show()
```

```python
# radius keyword for increase pie-chart size

plt.pie(expense_money, labels=expense_labels, radius=2)
plt.show()
```

```python
plt.axis("equal")
plt.pie(expense_money, labels=expense_labels, radius=5)
plt.show()
```

```python
# autopct keyword show each label in %, it is a string format

plt.axis('equal')
plt.pie(expense_money, labels=expense_labels, radius=1.5, autopct='%0.1f%%')
plt.show()
```

```python
# shadow keyword show shadow

plt.axis('equal')

plt.pie(expense_money, labels=expense_labels, shadow=True)

plt.show()
```

```python
# here expense_labels array have 5 value so we take 5 value in explode list

plt.pie(expense_money, labels=expense_labels, explode=[0, 0.1, 0, 0, 0.2])

plt.axis("equal")
plt.show()
```

```python
# color keyword it take equal element of labels

'r' - Red
'g' - Green
'b' - Blue
'c' - Cyan
'm' - Magenta
'y' - Yellow
'k' - Black
'w' - White
```

```python
array = [20, 25, 15, 10, 30]
mylabel = ['Math', 'Bangla', 'English', 'Biology', 'Chemistry']
mycolor = ['black', 'Red', 'Green', 'Cyan', 'Orange']

plt.pie(array, labels=mylabel, colors=mycolor)

plt.axis("equal")
plt.show()
```

```python
# startangle keyword rotate start element of expense_labels 0 to 360 degree

plt.pie(array, labels=mylabel, colors = mycolor ,startangle=90)

plt.axis("equal")
plt.show()
```

```python
plt.pie(array, labels=mylabel, colors=mycolor, startangle=180)

plt.axis("equal")
plt.show()
```

```python
plt.pie(array, labels=mylabel, colors=mycolor, startangle=270)

plt.axis("equal")
plt.show()
```
